```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df.isna().sum()
```

```python
# Select the features and target variable
x=df[['CreditScore','Age','Balance','EstimatedSalary']]
y=df['Exited']
print(x,y)
```

```python
# Split the train and test data
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
```

```python
x_test_original = x_test.copy()
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

#### Build ANN Model

#### Architecture

1)  Input Layer

2)  Hidden Layer 1

3)  Hidden Layer 2

4)  Output Layer

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model=Sequential([
    Dense(16,activation='relu',input_shape=(4,)),
    Dense(8,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

#### 16 neurons --> Learn patterns
#### 8 neurons --> Deeper Learning
#### Sigmoid --> Binary Classification

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

#### Adam --> Optimizer
#### Binary Crossentropy --> binary classification loss
#### Accuracy --> Evaluation metrics

```python
# Train the model
history=model.fit(x_train,y_train,epochs=20,validation_split=0.2,batch_size=32) # Use 20% of data for validation
```

#### Epoch --> One learning cycle
#### Batch size --> Rows processed together

```python
# Model Evaluation
loss,acc=model.evaluate(x_test,y_test)
print(f"Loss value : {loss}")
print(f"Accuracy : {acc}")
```

```python
pred_probs = model.predict(x_test)

for i in range(len(pred_probs)):
    if pred_probs[i][0] > 0.5:
        print("Probability:", pred_probs[i][0])
        print(x_test_original.iloc[i])
        break
```

```python
# Predictions
pred=model.predict(x_test)

# Convert probabilities into predictions
pred=(pred>0.5)
print(pred[:10])
```

#### 1 --> True
#### 0 --> False

#### Conclusion
    The above model predicts whether customer is willing to stay or not in an organization. 
    Determined using exited variable in the dataset.(0->Donot exit,1->Exits)
